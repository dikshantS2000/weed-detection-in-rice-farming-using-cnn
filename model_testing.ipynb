```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import os

# Load trained model
model = load_model("./rice_weed_classifier_mobilenet.keras")

# Image size
img_height, img_width = 224, 224

# Image path (use raw string for Windows)
img_path = r"./riceTest.jpg"

# Load and preprocess image
img = image.load_img(img_path, target_size=(img_height, img_width))
img_array = image.img_to_array(img)
img_array = img_array / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Predict
prediction = model.predict(img_array)
prediction_value = float(prediction.squeeze())  # Ensure it's a scalar

# Determine label (ensure it's a string)
label = str("Weed" if prediction_value > 0.5 else "Rice")

# Debug print to verify values
print(f"Prediction value: {prediction_value}")
print(f"Label: {label} (type: {type(label)})")

# Plot
plt.imshow(img)
plt.title(f"Prediction: {label} ({prediction_value:.2f})")  
plt.axis("off")
plt.show()
```
